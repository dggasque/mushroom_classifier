```python
import pandas as pd

df = pd.read_csv('mushrooms_mapped.csv')
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, random_state=42)
train, val = train_test_split(train, random_state=42)

train.shape, val.shape, test.shape
```

```python
target = 'class'
features = ['cap_surface', 'bruises', 'gill_spacing', 'gill_size', 'stalk_shape',
       'stalk_root', 'ring_type', 'spore_print_color', 'habitat']
X_train = train[features]
y_train = train[target]
X_val = val[features]
y_val = val[target]
```

```python
import category_encoders as ce
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier

X = df[features]
y = df[target]

# Use Ordinal Encoder, outside of a pipeline
encoder = ce.OrdinalEncoder()
X_encoded = encoder.fit_transform(X)

model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
model.fit(X_encoded, y)
```

```python
# Use Pdpbox
%matplotlib inline
import matplotlib.pyplot as plt
from pdpbox import pdp

feature = 'cap_surface'
for item in encoder.mapping:
    if item['col'] == feature:
        feature_mapping = item['mapping']
        
feature_mapping = feature_mapping[feature_mapping.index.dropna()]
category_names = feature_mapping.index.tolist()
category_codes = feature_mapping.values.tolist()

pdp_dist = pdp.pdp_isolate(model=model, dataset=X_encoded, model_features=features, feature=feature)
pdp.pdp_plot(pdp_dist, feature)

# Automatically change the xticks labels
plt.xticks(category_codes, category_names);
```

```python
# Import module for PDD interactions
from pdpbox.pdp import pdp_interact, pdp_interact_plot
```

```python
feature = 'ring_type'
for item in encoder.mapping:
    if item['col'] == feature:
        feature_mapping = item['mapping']
        
feature_mapping = feature_mapping[feature_mapping.index.dropna()]
category_names = feature_mapping.index.tolist()
category_codes = feature_mapping.values.tolist()
```

```python
feature_mapping.values.tolist()
```

```python
feature_mapping.index.tolist()
```

```python
feature = 'spore_print_color'
for item in encoder.mapping:
    if item['col'] == feature:
        feature_mapping = item['mapping']
        
feature_mapping1 = feature_mapping[feature_mapping.index.dropna()]
category_names1 = feature_mapping1.index.tolist()
category_codes1 = feature_mapping1.values.tolist()
```

```python
pdp_features = ['spore_print_color', 'ring_type']

interaction = pdp_interact(
    model=model, 
    dataset=X_encoded, 
    model_features=X_encoded.columns, 
    features=pdp_features
)


pdp = interaction.pdp.pivot_table(
    values='preds', 
    columns=pdp_features[0], # First feature on x axis
    index=pdp_features[1]    # Next feature on y axis
)[::-1]  # Reverse the index order so y axis is ascending

pdp = pdp.rename(
    index=dict(zip(category_codes, category_names)),
    columns=dict(zip(category_codes1, category_names1))
)
plt.figure(figsize=(10,8))
sns.heatmap(pdp, annot=True, fmt='.2f', cmap='viridis')
plt.title('Partial Dependence of Mushroom Edibility, on Ring Type & Spore Print Color');
```

```python
import plotly.graph_objs as go

surface = go.Surface(
    x=pdp.columns,
    y=pdp.index,
    z=pdp.values
)

fig = go.Figure(surface)
fig.update_layout(scene = dict(
                    xaxis_title='Spore Print Color',
                    yaxis_title='Ring Type',
                    zaxis_title='Predicted Probability of Poisonous Mushroom'),
                    width=700,
                    margin=dict(r=20, b=10, l=10, t=10))

fig.show()
```

```python
from joblib import dump
dump(pdp, 'pdp.joblib', compress=True)
```
